```python
# A.I SUM
# AIOT product
# CNC cumsum
# else 0
def custm_fn(x1,*l1):
    if x1 == 'A.I':
        return np.sum(l1)
    elif x1 == 'AIOT':
        return np.product(l1)
    elif x1 == 'CNC':
        return sum(np.cumsum(l1))
    else:
        return 0
```

```python
df[['Course','c1','c2','c3','c4']].apply(lambda x:custm_fn(*x),axis=1)
```

```python
df['Check Validation'] = df[['Course','c1','c2','c3','c4']].apply(lambda x:custm_fn(*x),axis=1)
```
