```python
df1 = pd.DataFrame({"Customer CNIC":np.random.randint(1000,2000,size=(20)),
                   "Cus Account":np.random.randint(1000000,2000000,size=(20))})
df2 = pd.DataFrame({"CNIC":np.random.randint(1000,2000,size=(20)),
                    "CustomerName":pd.util.testing.rands_array(10, 20),
                   "Account No":np.random.randint(1000000,2000000,size=(20)),
                   "DOB":pd.date_range(start='1/1/1990', periods=20),
                   "Gender":np.random.choice(["Male","Female"],size=20)})
df1
```

```python
df2
```

```python
df1.merge(df2,left_on=["Customer CNIC","Cus Account"],
          right_on=["CNIC","Account No"],how='outer')
```

```python
c = pd.DataFrame(df2.CNIC.value_counts())
c.columns = ["Tr Counts"]
c
```

```python
df1.merge(c, left_on="Customer CNIC", right_index=True,how='left')
```

```python
df1.join(c, left_on="Customer CNIC", right_index=True,how='left')
```
